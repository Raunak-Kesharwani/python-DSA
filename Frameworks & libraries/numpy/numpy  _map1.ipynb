```python
# dot product between two rows 
def dot1(a,b):
    s = 0 
    for i in range(len(a)):
        s += a[i]*b[i]
    return s

dot1([2,3],[3,4])
```

```python
def dot(a, b):
    return np.sum(a * b)
dot(np.array([2,3]),np.array([3,4]))
```

```python
def Mse(y , y_pred):
    assert type(y) == np.ndarray,  "not matches required datatype"

    loss = ((y - y_pred)**2)/y.shape[0]
    return loss 

Mse(np.random.randn(5) , np.random.randn(5))
```

```python
# ye mera wala h 
def Softmax(Z: np.array):
    soft = np.array([])
    for i in range(len(Z)):
        num = np.exp(Z[i])
        den = 0
        for j in np.concat((Z[0:i], Z[i+1:])):
            den += np.exp(j)
        den += num  

        soft = np.append(soft, num / den)

    return soft
```

```python
z = np.array([1,2,3,4,5])
Softmax(z)
```

```python
# AI
def softmax_loop(Z: np.ndarray):
    soft = np.zeros(len(Z))
    den = np.sum(np.exp(Z))

    for i in range(len(Z)):
        soft[i] = np.exp(Z[i]) / den

    return soft
softmax_loop(z)
```

```python
# AI
def softmax(Z: np.ndarray):
    Z = Z - np.max(Z)          # numerical stability
    exp_Z = np.exp(Z)
    return exp_Z / np.sum(exp_Z)
softmax(z)
```
