#### 19. Create a 8x8 matrix and fill it with a checkerboard pattern (★☆☆)

````python
a = np.ones((8,8))
a[1::2] = 0 
a
# it says select from first column than till last 2 specifies the jump 
````

````python
a = np.ones((8,8))
a[: , 1::2] = 0 
a
````

````python
a = np.ones((8,8) , dtype = int)
a[1::2 , 0::2] = 0 
a[0::2 , 1::2] = 0 
a
````

````python
Z = np.ones((8, 8), dtype=int)

# Set alternating cells to 0
Z[::2, ::2] = 0
Z[1::2, 1::2] = 0
Z
````

````python


# The original array 'a' already has the desired pattern. 
# Assuming you wanted to *change* an existing list of zeros into this pattern:

# Example: Starting with a list of lists of zeros
a = [[0.] * 8 for _ in range(8)]

# Equivalent of a[1::2 , 0::2] = 1
# Loop through rows starting at index 1, step 2
for row_idx in range(1, 8, 2):
    # Loop through columns starting at index 0, step 2
    for col_idx in range(0, 8, 2):
        a[row_idx][col_idx] = 1.0

# Equivalent of a[0::2 , 1::2] = 1
# Loop through rows starting at index 0, step 2
for row_idx in range(0, 8, 2):
    # Loop through columns starting at index 1, step 2
    for col_idx in range(1, 8, 2):
        a[row_idx][col_idx] = 1.0

# Print the resulting list of lists
for row in a:
    print(row)

````
